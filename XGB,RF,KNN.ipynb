```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix
import joblib

# ============================================
# SIMPLER Random Forest - Real-World Focus
# ============================================
rf = RandomForestClassifier(
    n_estimators=50,          #  Reduce from 200 (less overfitting)
    max_depth=10,             #  Reduce from 20 (shallower trees)
    min_samples_split=50,     #  Increase from 10 (more conservative splits)
    min_samples_leaf=20,      #  Prevent overfitting on small groups
    max_features='sqrt',      #  Limit features per tree (more diversity)
    max_samples=0.7,          #  Bootstrap with 70% samples
    class_weight="balanced",
    n_jobs=-1,
    random_state=42
)

rf.fit(X_train_scaled, y_train)

y_pred_rf = rf.predict(X_test_scaled)

print("=== Simpler Random Forest (Better Generalization) ===")
print(classification_report(y_test, y_pred_rf, digits=4))
print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred_rf))

# Calculate real-world metrics
tn, fp, fn, tp = confusion_matrix(y_test, y_pred_rf).ravel()
fpr = fp / (fp + tn) * 100
fnr = fn / (fn + tp) * 100

print(f"\n REAL-WORLD IMPACT:")
print(f"   False Positive Rate: {fpr:.3f}% ({fp:,} false alarms)")
print(f"   False Negative Rate: {fnr:.3f}% ({fn:,} missed attacks)")

joblib.dump(rf, "rf_nids_generalized.pkl")
```

```python
from xgboost import XGBClassifier
from sklearn.metrics import classification_report, confusion_matrix
import joblib

# ============================================
# SIMPLER XGBoost - Better Generalization
# ============================================
xgb = XGBClassifier(
    n_estimators=100,        #  Reduce from 300 (less overfitting)
    max_depth=4,             #  Reduce from 8 (simpler trees)
    learning_rate=0.1,       #  Increase from 0.05 (faster, less precise fitting)
    subsample=0.7,           #  More randomness
    colsample_bytree=0.7,    #  More randomness
    min_child_weight=5,      #  Prevent overfitting on small groups
    gamma=0.1,               #  Regularization (prune trees)
    reg_alpha=0.1,           #  L1 regularization
    reg_lambda=1.0,          #  L2 regularization
    scale_pos_weight=(y_train.value_counts()[0] / y_train.value_counts()[1]),
    eval_metric="logloss",
    tree_method="hist",
    random_state=42
)

# Train with early stopping
xgb.fit(
    X_train_scaled, 
    y_train,
    eval_set=[(X_test_scaled, y_test)],
    early_stopping_rounds=10,  # Stop if no improvement
    verbose=False
)

y_pred_xgb = xgb.predict(X_test_scaled)

print("=== Simpler XGBoost (Better Generalization) ===")
print(classification_report(y_test, y_pred_xgb, digits=4))
print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred_xgb))

joblib.dump(xgb, "xgb_nids_generalized.pkl")
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, confusion_matrix, precision_recall_curve, roc_curve, roc_auc_score
import matplotlib.pyplot as plt
import numpy as np
import joblib

# ============================================
# SIMPLER KNN - Real-World Focus
# ============================================
print(" Training KNN (Simplified for Generalization)...\n")

knn = KNeighborsClassifier(
    n_neighbors=15,           #  Increase from 7 (less overfitting, smoother decision boundary)
    weights="uniform",        # Change from "distance" (less sensitive to training data)
    metric="euclidean",       # Simpler metric than minkowski
    n_jobs=-1
)

knn.fit(X_train_scaled, y_train)

y_pred_knn = knn.predict(X_test_scaled)

print("=== Simpler KNN (Better Generalization) ===")
print(classification_report(y_test, y_pred_knn, digits=4))
print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred_knn))

# ============================================
# REAL-WORLD METRICS EVALUATION
# ============================================
print("\n" + "="*70)
print(" KNN - REAL-WORLD METRICS")
print("="*70)

# Get prediction probabilities
y_pred_proba_knn = knn.predict_proba(X_test_scaled)[:, 1]

# Calculate metrics from confusion matrix
tn, fp, fn, tp = confusion_matrix(y_test, y_pred_knn).ravel()

# Real-world metrics
fpr = fp / (fp + tn) * 100  # False Positive Rate
fnr = fn / (fn + tp) * 100  # False Negative Rate
precision_attack = tp / (tp + fp) * 100
recall_attack = tp / (tp + fn) * 100
specificity = tn / (tn + fp) * 100
f1_attack = 2 * (precision_attack * recall_attack) / (precision_attack + recall_attack)
auc_score = roc_auc_score(y_test, y_pred_proba_knn)

print(f"\n CONFUSION MATRIX BREAKDOWN:")
print(f"   True Negatives (TN):  {tn:,}  (Correctly identified Normal)")
print(f"   False Positives (FP): {fp:,}  (Normal flagged as Attack)")
print(f"   False Negatives (FN): {fn:,}  (Attack missed)")
print(f"   True Positives (TP):  {tp:,}  (Correctly identified Attack)")

print(f"\n FALSE ALARM ANALYSIS:")
print(f"   False Positive Rate: {fpr:.3f}%")
print(f"    Out of {tn+fp:,} normal flows, {fp:,} wrongly flagged as attacks")
print(f"    In production: {fp:,} false alarms would occur!")

print(f"\n MISSED ATTACK ANALYSIS:")
print(f"   False Negative Rate: {fnr:.3f}%")
print(f"    Out of {tp+fn:,} attacks, {fn:,} were MISSED")
print(f"    Security risk: {fn:,} attacks went undetected!")

print(f"\n PERFORMANCE METRICS:")
print(f"   Attack Precision:  {precision_attack:.2f}% (How many flagged attacks are real)")
print(f"   Attack Recall:     {recall_attack:.2f}% (How many real attacks were caught)")
print(f"   Attack F1-Score:   {f1_attack:.2f}%")
print(f"   Specificity:       {specificity:.2f}% (Normal traffic correctly identified)")
print(f"   ROC-AUC Score:     {auc_score:.4f}")

print(f"\n PRODUCTION IMPACT (per 1 million flows):")
flows_per_million = 1_000_000
normal_flows = int(flows_per_million * 0.83)  # ~83% normal
attack_flows = int(flows_per_million * 0.17)  # ~17% attacks

fp_per_million = int(normal_flows * (fpr/100))
fn_per_million = int(attack_flows * (fnr/100))

print(f"   Expected false alarms: {fp_per_million:,} per 1M flows")
print(f"   Expected missed attacks: {fn_per_million:,} per 1M flows")

# ============================================
# VISUALIZATION - 3 PLOTS
# ============================================
precision, recall, pr_thresholds = precision_recall_curve(y_test, y_pred_proba_knn)
fpr_curve, tpr_curve, roc_thresholds = roc_curve(y_test, y_pred_proba_knn)

plt.figure(figsize=(15, 5))

# Plot 1: Precision-Recall Curve
plt.subplot(1, 3, 1)
plt.plot(recall, precision, marker='.', linewidth=2, color='blue')
plt.xlabel('Recall (Attack Detection Rate)', fontsize=11)
plt.ylabel('Precision (Attack Accuracy)', fontsize=11)
plt.title('Precision-Recall Curve\nKNN', fontsize=12, fontweight='bold')
plt.grid(True, alpha=0.3)
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])

# Add current operating point
current_recall = recall_attack / 100
current_precision = precision_attack / 100
plt.scatter([current_recall], [current_precision], 
           color='red', s=100, zorder=5, label=f'Current (thresh=0.5)')
plt.legend()

# Plot 2: ROC Curve
plt.subplot(1, 3, 2)
plt.plot(fpr_curve, tpr_curve, linewidth=2, color='green', 
         label=f'KNN (AUC = {auc_score:.4f})')
plt.plot([0, 1], [0, 1], 'k--', label='Random Classifier')
plt.xlabel('False Positive Rate', fontsize=11)
plt.ylabel('True Positive Rate (Recall)', fontsize=11)
plt.title('ROC Curve\nKNN', fontsize=12, fontweight='bold')
plt.legend()
plt.grid(True, alpha=0.3)

# Plot 3: Threshold Analysis
plt.subplot(1, 3, 3)

# Test different thresholds
thresholds_test = np.arange(0.3, 0.95, 0.05)
fpr_list = []
fnr_list = []

for thresh in thresholds_test:
    y_pred_thresh = (y_pred_proba_knn >= thresh).astype(int)
    tn_t, fp_t, fn_t, tp_t = confusion_matrix(y_test, y_pred_thresh).ravel()
    
    fpr_t = fp_t / (fp_t + tn_t) * 100
    fnr_t = fn_t / (fn_t + tp_t) * 100
    
    fpr_list.append(fpr_t)
    fnr_list.append(fnr_t)

plt.plot(thresholds_test, fpr_list, marker='o', label='FPR (False Alarms)', 
         color='red', linewidth=2)
plt.plot(thresholds_test, fnr_list, marker='s', label='FNR (Missed Attacks)', 
         color='blue', linewidth=2)
plt.axvline(x=0.5, color='gray', linestyle='--', alpha=0.5, label='Default (0.5)')
plt.xlabel('Classification Threshold', fontsize=11)
plt.ylabel('Error Rate (%)', fontsize=11)
plt.title('Threshold Impact on Errors\nKNN', fontsize=12, fontweight='bold')
plt.legend()
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# ============================================
# THRESHOLD TUNING TABLE
# ============================================
print("\n" + "="*70)
print(" THRESHOLD TUNING - Optimize for Your Use Case")
print("="*70)
print(f"{'Threshold':<12} {'FPR %':<10} {'FNR %':<10} {'FP Count':<14} {'FN Count':<14}")
print("-" * 70)

for thresh in [0.3, 0.4, 0.5, 0.6, 0.7, 0.8, 0.9]:
    y_pred_thresh = (y_pred_proba_knn >= thresh).astype(int)
    tn_t, fp_t, fn_t, tp_t = confusion_matrix(y_test, y_pred_thresh).ravel()
    
    fpr_t = fp_t / (fp_t + tn_t) * 100
    fnr_t = fn_t / (fn_t + tp_t) * 100
    
    print(f"{thresh:<12.1f} {fpr_t:<10.3f} {fnr_t:<10.3f} {fp_t:<14,} {fn_t:<14,}")

print("\n THRESHOLD RECOMMENDATIONS:")
print("   0.3-0.4: High sensitivity (catch ALL attacks, many false alarms)")
print("   0.5-0.6:  BALANCED (recommended for most cases)")
print("   0.7-0.8: Low false alarms (production-friendly, may miss some attacks)")
print("   0.9+:    Very conservative (minimal alarms, risk missing attacks)")

# ============================================
# COMPARE WITH DIFFERENT K VALUES
# ============================================
print("\n" + "="*70)
print(" K-VALUE IMPACT ANALYSIS")
print("="*70)
print(f"{'K Value':<10} {'Accuracy %':<12} {'FPR %':<10} {'FNR %':<10} {'Train Time (s)':<15}")
print("-" * 70)

import time

k_values = [3, 5, 7, 10, 15, 20, 25, 30]

for k in k_values:
    start_time = time.time()
    
    knn_temp = KNeighborsClassifier(
        n_neighbors=k,
        weights="uniform",
        metric="euclidean",
        n_jobs=-1
    )
    knn_temp.fit(X_train_scaled, y_train)
    
    train_time = time.time() - start_time
    
    y_pred_temp = knn_temp.predict(X_test_scaled)
    tn_t, fp_t, fn_t, tp_t = confusion_matrix(y_test, y_pred_temp).ravel()
    
    accuracy = (tp_t + tn_t) / (tp_t + tn_t + fp_t + fn_t) * 100
    fpr_t = fp_t / (fp_t + tn_t) * 100
    fnr_t = fn_t / (fn_t + tp_t) * 100
    
    marker = "" if k == 15 else ""
    print(f"{k:<10} {accuracy:<12.2f} {fpr_t:<10.3f} {fnr_t:<10.3f} {train_time:<15.2f} {marker}")

print("\n K-VALUE RECOMMENDATIONS:")
print("   k=3-7:   High accuracy but OVERFITS (poor real-world performance)")
print("   k=10-15:  BALANCED (good generalization)")
print("   k=20-30: More robust but slower (consider for production)")

# ============================================
# SAVE MODEL
# ============================================
joblib.dump(knn, "knn_nids_generalized.pkl")
print("\n Model saved: knn_nids_generalized.pkl")

print("\n" + "="*70)
print(" KNN EVALUATION COMPLETE!")
print("="*70)
```
